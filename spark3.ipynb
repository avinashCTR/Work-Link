```scala
//   val boostScore: UserDefinedFunction = udf{ (vertical: String, is_category: Boolean, is_pure_brand: Boolean) => 
//     var is_category_score = 1
//     var is_category_score = 1
//     if(is_category){
//         is_category_score = 3
//     }
//     if(is_pure_brand){
//         is_pure_brand_score = 2
//     }
      
//       myMap(vertical)*is_category_score* is_category_score
//   }
import org.apache.spark.sql.functions._
import org.apache.spark.sql.expressions.UserDefinedFunction
val myMap: Map[String, Int] = Map("Electronics" -> 7, "Groceries" -> 20,  "Beauty" -> 6, "Fashion"-> 10, "Home & Lifestyle"->6, "OtherVerticals"->6)

val boostScore: UserDefinedFunction = udf { 
  (vertical: String, is_category: Boolean, is_pure_brand: Boolean) => 
    var is_category_score = 1
    var is_pure_brand_score = 1  // Fixed variable name

    if (is_category) {
        is_category_score = 3
    }
    if (is_pure_brand) {
        is_pure_brand_score = 2
    }
    
    // Ensure vertical exists in myMap to avoid KeyError
    myMap.getOrElse(vertical, 1) * is_category_score * is_pure_brand_score
}

```
